```python
#process payrate
df['payrate']
```

```python
#same operation as a fuction ..pass a new data frame and return another
def paysplit_func(dfi):
    #split at -
    pay_split = df['payrate'].str[0:-1].str.split('-', expand=True)
    #for col 0 preprocessing
    #remove space in left and right 
    pay_split[0] =  pay_split[0].str.strip()
    #remove comma 
    pay_split[0] = pay_split[0].str.replace(',', '')
    #remove all character in two condition
    # 1 remove if only character
    # 2 if start in number remove after all character
    pay_split[0] = pay_split[0].str.replace(r'\D.*', '')
    #display 
   # pay_split[0].head()
    #convert to number, renamecolumns as min_pay and max_pay null values become NaN
    pay_split[0] = pd.to_numeric(pay_split[0], errors='coerce')
    pay_split[1] = pd.to_numeric(pay_split[1], errors='coerce')
    pay=pd.concat([pay_split[0], pay_split[1]], axis=1, sort=False)
    pay.rename(columns={0:'min_pay', 1:'max_pay'}, inplace=True)
    #pay.head()
    #create a new data frame df1 and return
    dfx=pd.concat([df, pay], axis=1, sort=False)
    dfx.head()
    return dfx
```

```python
df1=paysplit_func(df)
df1
```

```python
#process experience
df['experience']
```

```python
# do preprocessing for experience
experience_split = df['experience'].str[0:-1].str.split('-', expand=True)
experience_split.head()
#remove space in left and right 
experience_split[1] =  experience_split[1].str.strip()
#remove comma 
experience_split[1] = experience_split[1].str.replace('yr', '')
#remove all character in two condition
# 1 remove if only character
# 2 if start in number remove after all character
experience_split[1] = experience_split[1].str.replace(r'yr', '')
#display 
experience_split[1].head()
experience_split[0] = pd.to_numeric(experience_split[0], errors='coerce')
experience_split[1] = pd.to_numeric(experience_split[1], errors='coerce')
experience=pd.concat([experience_split[0], experience_split[1]], axis=1, sort=False)
experience.rename(columns={0:'min_experience', 1:'max_experience'}, inplace=True)
experience.head()
```

```python
#concate it in df2
df2=pd.concat([df1, experience], axis=1, sort=False)
df2.head()
```

```python
#create new data frame with min exp, min pay under a particular job title.
#objective is to predict unknown ay rates with experiences
daf=pd.concat([df2.min_experience,df2.min_pay,df2.jobtitle], axis=1, sort=False)
daf1=daf.loc[daf.jobtitle=="Java Developer"]
daf1.isnull().sum()
```

```python
daf1=daf1.dropna()
daf1
```

```python
sns.stripplot(x='min_experience', y='min_pay', data=daf1, jitter=True)
```

```python

#create an np array for polynomialprediction
aaa=daf1.index
l=len(aaa)
X=[]
X=[0 for i in range(l)]
Y=[]
Y=[0 for i in range(l)]
j=0
for i in range(l-1):
     X[j]=daf1.loc[aaa[i],'min_experience']
     Y[j]=daf1.loc[aaa[i],'min_pay']
     j=j+1
X1=np.array(X).reshape(-1,1)
Y1=np.array(Y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X1,Y1)
plt.show()
```

```python
#create a model and predict
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(X1)
X1_ = transformer.transform(X1)
#fit the model
model = LinearRegression().fit(X1_,Y1)
r_sq = model.score(X1_,Y1)
print("accuracy:",r_sq)
y_pred = model.predict(X1_[0].reshape(1,-1))
print("predicted value for", X1_[0], "is..",y_pred)
```
