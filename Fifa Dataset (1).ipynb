# Manipulating and Plotting Dataset of Football Players

```python
#Loading Libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file=pd.read_csv(r"C:\Users\hello\OneDrive\Desktop\players_20.csv")
```

```python
#Finding how many colums in dataset
for col in file.columns:
    print(col)
```

```python
#Finding the nationality count from dataset
file['nationality'].value_counts()
```

```python
#Finding the nationality count of first 10
file['nationality'].value_counts()[0:10]
```

```python
#To extract names of top 5 countires
file['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(file['nationality'].value_counts()[0:5].keys()),list(file['nationality'].value_counts()[0:5]), color='g')
plt.show()
```

```python
#Getting player wages with names
player_salary=file[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
#Sorting player salaries
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
#Outputting sorted values
player_salary.head()
```

```python
#Plotting bar graph based on salaries
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=['blue','red','green','pink', 'yellow'])
plt.show()
```
